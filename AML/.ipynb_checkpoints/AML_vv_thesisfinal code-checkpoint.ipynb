Preprocessing of the dataset

```python
def preprocess_data(odata, clean=True):
    """ Cleans, filters and transforms the given data.
        Data cleaning:
            - Report the number of transactions with Null ID and remove them.
            - Report the number of transactions with Amount <= €0.0 and remove them.
            - Report the number of transactions with €0.0 < Amount < €1.0 and remove them.
            - Report the number of duplicate transactions and remove them.
        Data transformation:
            - Replace Null numerical columns to 0.0.
            - Convert column ID to string.
            - Convert column CU_LEN to new column Entity_Type (ToDo).
    """
    data = odata
    columns_to_display = ['ID', 'TIMESTAMP', 'AMOUNT', 'CLIENT_USERNAME_MASKED', 'DEBIT_IBAN_MASKED', 'RECEIVER_ACC_IBAN_MASKED']
    show_rows = 5
    columns_to_flag = ['DEBIT_IBAN_MASKED', 'RECEIVER_ACC_IBAN_MASKED']
    
    
    
    # Report the number of transactions with Null ID and remove them.
    num = count_null(data['ID'])
    if num > 0:
        print("The following {0:} transaction{1:} ha{2:} Null Unique ID{1:}:".format(num, "s" if num > 1 else "", "ve" if num > 1 else "s"))
        display(data[data['ID'].isna()][columns_to_display].head(show_rows))

    
    # Report the number of transactions with 0.0 amount and remove them.
    num = len(data[data['AMOUNT'] <= 0.0])
    if num > 0:
        print("The following {0:} transaction{1:} ha{2:} €0.0 or less Amount{1:}:".format(num, "s" if num > 1 else "", "ve" if num > 1 else "s"))
        display(data[data['AMOUNT'] <= 0.0][columns_to_display].head(show_rows))
    
    # Report the number of transactions with 0.0<amount<1.0 and remove them.
    num = len(data[data['AMOUNT'] < 1.0])
    if num > 0:
        print("The following {0:} transaction{1:} ha{2:} Amount{1:} less than €1.0:".format(
            num, "s" if num > 1 else "", "ve" if num > 1 else "s"))
        display(data[data['AMOUNT'] < 1.0][columns_to_display].head(show_rows))
        
    # Report the number of duplicate transactions and remove them.
    non_dups = data.drop_duplicates()
    num = len(data) - len(non_dups)
    if num > 0:
        print("The following {} duplicate transaction{} {} found:".format(num, "s" if num > 1 else "", "were" if num > 1 else "was"))
        display(data[~data.duplicated(keep=False)][columns_to_display].sort_values('ID', kind='mergesort').head(show_rows))

    if clean:
        # Remove duplicates
        data = non_dups
        # Remove Null IDs
        data = data[~data['ID'].isna()]
        
        # Remove all the low values on AMOUNT
        data = data[data['AMOUNT'] >= 1.0]
        
        # remode the usernames not found in the dataset
        if len(data[data['CU_LEN'].isna()]):
            print("Some Usernames were not found in the file. Setting their length to 0.")
        
        data['CU_LEN'] = data['CU_LEN'].fillna(0)   
    
        # encode CU_LEN to words
        data.CU_LEN = data.CU_LEN.apply(lambda x: ('individuals' if x < 10 else 'legal entity'))
        
        # choose only the numerical values
        numerical_columns = data.select_dtypes(['float', 'int'])
        for col in numerical_columns:
            data[col] = data[col].fillna(0.0)
            
        # Convert dates to datetime type
        data['DATE'] = pd.to_datetime(data.TIMESTAMP, errors='coerce').dt.date
        
        #convert object type to float
        data["ETE_COMISSION_OUT"] = data.ETE_COMISSION_OUT.astype(float)
        data["URGENT_EXPENCES_OUT"] = data.URGENT_EXPENCES_OUT.astype(float)
        data["EXCHANGE_PROFIT_OUT"] = data.EXCHANGE_PROFIT_OUT.astype(float)
        data["DEPT_EXPENCES_OUT"] = data.DEPT_EXPENCES_OUT.astype(float)
        
        # convert the countries
        data['DEST_COUNTRY'] = data.DEST_COUNTRY.str[3:].str.upper()
        data['DEST_COUNTRY'] = data['DEST_COUNTRY'].replace('ΕΛΛΑΔΑ', 'GREECE')

    print('Returned {:,d} clean transactions from {:,d} input ones.'.format(len(data), len(odata)))
    return data

# preprocess the dataset
data = preprocess_data(data, clean = True)
```

# Aggregations

```python
def tranfers_accounts(data, country = None):
    ''' Does aggregations per account based on the amount.
    
    Arguments:
        data {Dataframe} -- The dataset.
        country {list of strings} -- county values in the corresponding column of the dataframe
       
    Returns:
        {Dataframe} -- Joinned Dataframe of the accounts containing aggregations of the 
        amount and each account's balance
    '''
    
    # checks the country
    if country != None: data = data.loc[data.DEST_COUNTRY.isin(country)]
    
    # group by Debit account based on the amount sorted by count
    debit_group = data.groupby(by = 'DEBIT_IBAN_MASKED')['AMOUNT'].\
                                agg(['count', 'sum', 'mean']).\
                                sort_values(by = 'count', ascending = False).\
                                reset_index().\
                                rename(columns = {'DEBIT_IBAN_MASKED' : 'Account', \
                                                  'count' : 'Debit_count',\
                                                  'sum' : 'Debit_sum',\
                                                  'mean' : 'Debit_mean'})
    
    # group by Receiver Account based on the amount sorted by count
    receive_group = data.groupby(by = 'RECEIVER_ACC_IBAN_MASKED')['AMOUNT'].\
                                    agg(['count', 'sum', 'mean']).\
                                    sort_values(by = 'count', ascending = False). \
                                    reset_index().\
                                    rename(columns = {'RECEIVER_ACC_IBAN_MASKED' : 'Account',\
                                                      'count' : 'Receive_count',\
                                                      'sum' : 'Receive_sum', \
                                                      'mean' : 'Receive_mean'})
    # join the two dataframes
    joinned = pd.merge(debit_group, receive_group, how = 'outer')
    
    # find the columns that contain null values
    null_columns = joinned.isna().sum()[joinned.isna().sum()>0].index.tolist()
    
    # fill the null values
    for null_column in null_columns:
        joinned[null_column] = joinned[null_column].fillna(0.0)

    
    joinned['Balance_Account'] = joinned.Receive_sum - joinned.Debit_sum
    
    return joinned
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)


def username_to_accounts(data, aggregated_account, agg_function = 'count'):
    ''' Does aggregations per username for either debit or receiver account.
    
    Arguments:
        data {Dataframe} -- The dataset.
       
    Returns:
        {Dataframe} -- Aggregated result based on the AMOUNT and information of the CU_LEN sorted 
                        by the number of transactions.
    '''
      
    df = data.groupby(by = ['CLIENT_USERNAME_MASKED', aggregated_account],as_index = False). \
                             agg({'AMOUNT' : agg_function, 'CU_LEN' : 'first' })
    
    
    if type(agg_function) is list:
        
        # sort the dataframe
        sorted_df = df.sort_values(by = [('AMOUNT', 'count')], ascending = False)
        
        # remove the multiindex
        sorted_df.columns = ['_'.join(col) for col in sorted_df.columns.values]
        
        
    else: sorted_df = df.sort_values(by = 'AMOUNT', ascending = False)
        
    return sorted_df
```

```python
def info(data, characteristic, user):
    ''' Returns rows of the original Dataframe either choosing Username, Debit or Receiver account.
    
    Arguments:
        data {Dataframe} -- The dataset.
        characteristic {String} -- One of the three key columns of the dataframe
                       {possible values} -- 'CLIENT_USERNAME_MASKED', 'DEBIT_IBAN_MASKED', 'RECEIVER_ACC_IBAN_MASKED'
    Returns:
        {Dataframe} -- Rows of the dataset if characteristic == 'DEBIT_IBAN_MASKED', 'RECEIVER_ACC_IBAN_MASKED',
                        Added number of unique accounts if characteristic == 'CLIENT_USERNAME_MASKED'
    '''
    characteristics = ['CLIENT_USERNAME_MASKED', 'DEBIT_IBAN_MASKED', 'RECEIVER_ACC_IBAN_MASKED']
    
    if characteristic not in characteristics:
        
        error_char_string = ', '.join(['\''+characteristic+'\'' for characteristic in characteristics])
        raise ValueError('Characteristic not available. Try one between: ' + error_char_string)
    
    customer_transactions = data.loc[data[characteristic] == user]
    
    if characteristic == 'CLIENT_USERNAME_MASKED':
        
        unique_accounts_used = data.groupby('CLIENT_USERNAME_MASKED').apply(lambda x: len(x['DEBIT_IBAN_MASKED'].unique())).reset_index().rename(columns = {0 : 'no_debit_acc'})
        unique_accounts_given = data.groupby('CLIENT_USERNAME_MASKED').apply(lambda x: len(x['RECEIVER_ACC_IBAN_MASKED'].unique())).reset_index().rename(columns = {0 : 'no_receive_acc'})
 
        display(unique_accounts_used.loc[unique_accounts_used['CLIENT_USERNAME_MASKED'] == user])
        display(unique_accounts_given.loc[unique_accounts_used['CLIENT_USERNAME_MASKED'] == user])
        
    return customer_transactions
```

# Validity of the dataset

```python
from scipy.stats import chisquare
import math
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

def benford(data, prints = True):
    ''' Plots Benford's law against the dataset's distribution of first digit.
    
    Arguments:
        data {Series} -- A numeric column for checking Banford's law.
       
    Returns:
        plt_object -- The matlibplot object containing the histogram.
        {Dataframe} -- The statistic value and the p-value of the chi-square test.
    '''

    
    # turn the column to a dataframe
    data_df = data.to_frame()

    # multiply it by 100 in order to ensure keeping the first digit
    data_df['multiply_by_100'] = data * 100

    # keep only the first digit
    data_df['first_digit'] = data_df.multiply_by_100.astype(str).str[0]

    # find out the frequency of each first digit
    distribution = data_df['first_digit'].value_counts(normalize = True).to_frame()
    
    # width of the bars
    width = 0.40

    numbers = [float(n) for n in range(1, 10)]
    benford = [math.log10(1 + 1 / d) for d in numbers]

    fig = plt.figure(figsize = (16, 10))

    ax = plt.gca()
    
    # barplots
    
    bar1 = ax.bar(np.asarray(numbers) - width/2, benford, width, color = 'royalblue', label = 'Benford\'s law')
    bar2 = ax.bar(np.asarray(numbers) + width/2, distribution.first_digit, width, color = 'crimson', label = 'Test data')

    #plt.plot(np.asarray(numbers) - width/2, benford, color = 'darkorange')
   # plt.plot(np.asarray(numbers) + width/2, distribution.first_digit, color = 'steelblue')
    
    # annotate the bars
    
    for p in bar1.patches:
        ax.text(p.get_x() + p.get_width() / 2 - 0.05 ,p.get_height()+0.003, '{:.2f}%'.format(p.get_height()) , color='k', fontweight='bold', ha = "center")

    for p in bar2.patches:
        ax.text(p.get_x() + p.get_width() / 2 + 0.05 ,p.get_height()+0.003, '{:.2f}%'.format(p.get_height()) , color='k', fontweight='bold', ha = "center") 
    
    # set titles
    
    ax.set_ylabel('Frequency/Percentage', fontsize = 15)
    ax.set_xlabel('First digit', fontsize = 15)
    
    ax.set_xticks(numbers)
    ax.set_title('Benford\'s Law on Dataset', fontsize = 20, fontweight = 'bold')
    plt.legend()

    plt.show()
    
    # perform the chi-square test
    if prints:
        statistic, p_value = chisquare(distribution.first_digit, benford)

        df = pd.DataFrame({'Chi-Square Statistic' : '{:.5f}'.format(statistic), 'p-value' : '{:.5f}'.format(p_value)}, index = [0])
        display(df)
        
# Plot the Benford on these data   
benford(data.AMOUNT)
```

```python
def barplot_dates(data):
    
    # group the dates
    agg_date = data.groupby('DATE')['AMOUNT'].count().to_frame().reset_index().\
                                        rename(columns = {'AMOUNT' : 'Number_of_transactions'})
    
    # color palettes and ranking
    pal = sns.color_palette('Blues_d', len(agg_date))
    rank = agg_date['Number_of_transactions'].argsort().argsort() 
    
    # barplot
    fig = plt.figure(figsize = (16, 11))
    g = sns.barplot(x = 'DATE',y = 'Number_of_transactions',data = agg_date,\
                    palette = np.array(pal[::-1])[rank])
    
    # annotations
    for index, row in agg_date.iterrows():
        g.text(row.name, row.Number_of_transactions, "{0:,.0f}".format(row.Number_of_transactions),\
               color = 'black', fontweight = 'bold', ha = "center", fontsize = 12)
    
    # labels
    sns.despine()
    plt.ylabel('Number of transactions', fontsize = 15)
    plt.xlabel('Date', fontsize = 15)

    plt.xticks(rotation = 45)
    plt.title('Number of transactions per Date', fontsize = 20)

    fig.text(0.5, 1e-8, 'Number of transactions on each day', ha ='center', \
             fontsize = 18)

    plt.show()
    
# Plot the barplot 

barplot_dates(data)
```

```python
from scipy.interpolate import make_interp_spline, make_lsq_spline, BSpline

def plot_histogram(x_data, y_data, y_label, log_axis=False, show_pareto=False, verbose=True):
    
    """ Plots a histogram of the given data, the mean and median and the corresponding piecewise and interpolated
        trendlines.
    Arguments:
        x_data {DataFrame} -- The bins to use for the x axis.
        y_data {DataFrame} -- The values to group in the given bins.
        y_label {str} -- The name of the y data to display in the labels.
    Keyword Arguments:
        log_axis {bool} -- Indicates whether the y axis will be displayed in log scale or not (default: {False}).
        show_pareto {bool}  -- Indicates whether the Pareto cummulative percentage should be displayed (default: {False}).
        verbose {bool} -- Indicates whether short or long descriptions will be displayed (default: {True}).
    Returns:
        plt_object -- The matlibplot object containing the histogram.
        coverage -- The percentage of values that are included in the histogram's bins.
    """
    # Ensure the bins are ordered and all different.
    x_data = sorted(x_data)
    if set(x_data[1:]) & set(x_data[:-1]):  # Check the corresponding elements of x[0..n-1] and x[1..n].
        x_corrections = np.zeros(len(x_data))
        for itr in range(1, len(x_data)):  # Count consecutive equal numbers
            x_corrections[itr] = x_corrections[itr-1]+1 if x_data[itr] == x_data[itr-1] else 0
        jitter = min(1.0, (min(x_data)+max(x_data))/2.0)
        x_data = x_data + x_corrections*jitter
    x_data = sorted(x_data)
    
    # Plot the histogram
    hist_n, hist_bins, _ = plt.hist(y_data, bins=x_data, log=log_axis, color='steelblue', edgecolor='black')
    # Add the bar labels
    m_factor = 0.05 if log_axis else 0.01
    margins = [np.diff(hist_bins)/2.0, m_factor/plt.gcf().get_size_inches()[1]]  # Margins = (Bin's Center, % of Max Y)
    margins[1] = (np.log(max(hist_n)) if log_axis else max(hist_n)) * margins[1]
    axis = plt.gca()
    for itr in range(len(hist_bins)-1):
        axis.text(hist_bins[itr]+margins[0][itr], (1+m_factor)*hist_n[itr]+margins[1], "{0:,.0f}".format(hist_n[itr]),
                  color='k', fontweight='bold', ha="center")
        
    # Plot the mean and median values
    median, mean = np.median(y_data), np.mean(y_data)
    y1, y2 = max(hist_n)*(1-m_factor)**10, max(hist_n)  # Define the heights where the text will written.
    median_line = plt.plot((median, median), (0.0, (1-m_factor)*y1), 'r--', linewidth = 2,
                           label=("Median" + (" of {}" if verbose else "")).format(y_label))
    axis.text(median, (1-m_factor)**2*y1, "  Median: \u20ac{:,.2f}".format(median), color='r', fontweight='bold')
    mean_line = plt.plot((mean, mean), (0.0, (1-m_factor)*y2), '--', color='magenta', linewidth = 2,
                         label=("Mean" + (" of {}" if verbose else "")).format(y_label))
    axis.text(mean, (1-m_factor)**2*y2, "  Mean: \u20ac{:,.2f}".format(mean), color='magenta', fontweight='bold')
    
    # Plot a piecewise trendline by joining the bar heights
    trend_line = plt.plot(hist_bins[1:]-margins[0], np.maximum(hist_n, 0.001*margins[1]), color='c', label="Piecewise Linear"+(" Trendline" if verbose else ""))
    # Plot an interpolated trendline
    x_vals = hist_bins[1:] - margins[0]
    spl = make_interp_spline(x_vals, hist_n, k=2)  # BSpline object
    sample_x_vals = np.linspace(x_vals.min(), x_vals.max(), 100)  # The number of num_of_bins.
    trend_spline = plt.plot(sample_x_vals, np.maximum(spl(sample_x_vals), 0.001*margins[1]), color='darkorange', linewidth = 2,
                            label="Interpolated" + (" Trendline" if verbose else ""))
    
    # Transaction coverage percentage
    coverage = sum(hist_n) / len(y_data)
    cov_label = "Coverage: {:.2%}{}".format(coverage,
                ". Remaining Transactions: {:,.0f} out of {:,d}.".format(len(y_data)-sum(hist_n), len(y_data)) if coverage<1.0 else "")
    if coverage < 1.0:
        scale_f = 0.2/plt.gcf().get_size_inches()[0], 0.2/plt.gcf().get_size_inches()[1]
        rem_point = (max(max(hist_bins), hist_bins[-2]+margins[0][-2])*(1-scale_f[0]),
                     ((len(y_data)-sum(hist_n))*(1-scale_f[1]))*plt.gca().get_ylim()[1]/len(y_data))  # Scaled on YAxis.
        remaining = plt.scatter(rem_point[0], rem_point[1], color="dimgrey", marker="p", label="Remaining Transactions")
        axis.text(rem_point[0]*(1-scale_f[0]), rem_point[1]*(1+5*scale_f[1]), "{0:,.0f}".format(len(y_data)-sum(hist_n)),
                  color='dimgrey', fontweight='bold', ha="center")
        plt.plot(np.array([hist_bins[-1]-margins[0][-1], rem_point[0]]), np.array([hist_n[-1], rem_point[1]]), 'c--')
        plt.plot(np.append(sample_x_vals[-10:], rem_point[0]), np.append(spl(sample_x_vals[-10:]), rem_point[1]), '--',
                 color='darkorange', linewidth=2)
        
    # Plot Pareto Cummulative Percentage
    if show_pareto:
        pareto = hist_n.cumsum() / hist_n.sum()
        scale_p = 0.05 * plt.gca().get_ylim()[1]  # An arbitrary height to display the Pareto line.
        pareto_p = plt.plot(hist_bins[1:]+margins[0][itr], pareto+scale_p, color="lightgreen", marker="^", ms=5,
                            label=("Pareto Cumulative "+("Percentage" if verbose else "%")))
        for itr in range(1, len(hist_bins)-1):
            axis.text(hist_bins[itr]+margins[0][itr], pareto[itr]-5*margins[1]+scale_p, "{:.1%}".format(pareto[itr]),
                      color='darkgreen', fontweight='bold', ha="center", rotation=45)

    # Format the graph
    plt.title(("Histogram of " if verbose else "")+"{}".format(y_label), fontsize=20, fontweight='bold')
    bin_mean, bin_median = np.mean(np.diff(hist_bins[1:])), np.median(np.diff(hist_bins[1:]))
    plt.xlabel((("{0:} " if verbose else "Values ") + "in Bins of \u20ac{1:,.2f}{2:}\n{3:}").format(
                y_label, bin_mean, " (on average)" if bin_mean != bin_median else "", cov_label), fontsize=15, labelpad=10)
    plt.xticks(hist_bins, rotation=45)
    axis.set_xticklabels(["\u20ac{:,.0f}".format(a_bin) for a_bin in hist_bins])
    plt.xlim(min(hist_bins), max(hist_bins))
    plt.ylabel(("Number of " if verbose else "") + "Transactions{}".format(" (Log Scale)" if log_axis else ""), fontsize=15)
    plt.legend([trend_line[0], trend_spline[0], mean_line[0], median_line[0]]
               + ([pareto_p[0]] if show_pareto else []) + ([remaining] if len(y_data)>sum(hist_n) else []),
               (trend_line[0].get_label(), trend_spline[0].get_label(), mean_line[0].get_label(), median_line[0].get_label())
               + ((pareto_p[0].get_label(),) if show_pareto else tuple()) + ((remaining.get_label(),) if len(y_data)>sum(hist_n) else tuple()),
               shadow=True)
    return plt, coverage

def display_amount_graphs():
    
    """ Displays a composite graph of the transaction amounts and their commissions. """
    # Setup the layout
    grid = plt.GridSpec(3, 3, wspace=0.2, hspace=0.1)
    plt.figure(figsize=(16, 26))
    plt.minorticks_on()

    # Plot the histogram for AMOUNT
    plt.subplot(grid[0, 0:])
    bins = [0] + list(np.round(np.linspace(300, 20000, 20), -1).astype(int))
    plot_histogram(bins, data.AMOUNT, "Transfer Amount", log_axis=True, show_pareto=True, verbose=True)

    # For the commissions' sub-graphs, use bin [0, 0.01) for zero (non-existing) expenses.
    
    plt.figure(figsize=(16, 16))
    log_scale, num_of_bins = True, 5
    
    # Plot the histogram for ETE_COMISSION_OUT
    
    plt.subplot(grid[1, 0:1])
    bins = [0] + list(np.linspace(0.01, max(data.ETE_COMISSION_OUT), num_of_bins))
    plot_histogram(bins, data.ETE_COMISSION_OUT, "NBG Commission", log_axis=log_scale, verbose=False)
    
    # Plot the histogram for NONSTP_EXPENCES_OUT
    
    plt.subplot(grid[1, 1:2])
    bins = [0] + list(np.linspace(0.01, max(data.NONSTP_EXPENCES_OUT), num_of_bins))
    plot_histogram(bins, data.NONSTP_EXPENCES_OUT, "Non-STP Expenses", log_axis=log_scale, verbose=False)
    plt.ylabel("")  # Remove the repeated ylabel to free some space.
    # Plot the histogram for SUM_COMISSION_OUT
    plt.subplot(grid[1, 2:3])
    bins = [0] + list(np.linspace(0.01, max(data.SUM_COMISSION_OUT), num_of_bins))
    plot_histogram(bins, data.SUM_COMISSION_OUT, "Total Commission", log_axis=log_scale, verbose=False)
    plt.ylabel("")

    # Plot the histogram for DEPT_EXPENCES_OUT
    plt.figure(figsize=(16, 16))
    plt.subplot(grid[2, 0:1])
    bins = [0] + list(np.linspace(0.01, max(data.DEPT_EXPENCES_OUT), num_of_bins))
    plot_histogram(bins, data.DEPT_EXPENCES_OUT, "Debtor Expenses", log_axis=log_scale, verbose=False)
    # Plot the histogram for URGENT_EXPENCES_OUT
    plt.subplot(grid[2, 1:2])
    bins = [0] + list(np.linspace(0.01, max(data.URGENT_EXPENCES_OUT), num_of_bins))
    plot_histogram(bins, data.URGENT_EXPENCES_OUT, "Urgent Transfer Expenses", log_axis=log_scale, verbose=False)
    plt.ylabel("")  # Remove the repeated ylabel to free some space.
    # Plot the histogram for EXCHANGE_PROFIT_OUT
    plt.subplot(grid[2, 2:3])
    bins = [0] + list(np.linspace(0.01, max(data.EXCHANGE_PROFIT_OUT), num_of_bins))
    plot_histogram(bins, data.EXCHANGE_PROFIT_OUT, "Exchange Profit", log_axis=log_scale, verbose=False)
    plt.ylabel("")

    plt.show()

display_amount_graphs()
```

# Numerical Tests of distributions and Power Laws

```python
import warnings
warnings.filterwarnings("ignore")

from scipy.stats import skew, shapiro, anderson, kstest, normaltest, pareto

def statistics (data, prints = True):
    
    pd.set_option('display.float_format', '{:,.2f}'.format)
    
    mean, median, skewness = round(np.mean(data.AMOUNT),2), np.median(data.AMOUNT), skew(data.AMOUNT)

    statistic_shapiro, p_shapiro = shapiro(data.AMOUNT)

    statistic_kstest, p_kstest = kstest(data.AMOUNT, 'norm')

    statistic_Dagostino, p_Dagostino = normaltest(data.AMOUNT, axis = 0)
    
    pareto_smp = pareto(1.16)

    statistic_pareto, p_pareto = kstest(data.AMOUNT, pareto_smp.cdf)
    
    df = pd.DataFrame({'statistic': [statistic_shapiro, statistic_kstest, statistic_Dagostino, statistic_pareto],\
              'p_value': [p_shapiro, p_kstest, p_Dagostino, p_pareto]}).\
                rename(index = {0: 'Shapiro', 1: 'kstest', 2: 'D\'Agostino', 3: 'Pareto'})
    
    statistic_values = pd.DataFrame({'Mean' : [mean], 'Median' : [median], 'Skew' : [skewness]})
    
    if prints:
        display(df)
        display(statistic_values)

# call the function
statistics(data)
```

# Outliers

```python
from scipy import stats
import numpy as np

def find_outliers(data, range_ratio=0.75, group_methods=False):
    """ Performs outlier detection using several methods. See: http://colingorrie.github.io/outlier-detection.html

    Arguments:
        data {DataFrame} -- The data to examine

    Keyword Arguments:
        range_ratio {float} -- Defines the percentage of the data that will be considered as candidate outliers
                                by the Naive Range method (default: {0.75}).
        group_methods {bool} -- Indicates wheather the methods that produce the same outliers will be grouped together
                                (default: {False}).

    Returns:
        list -- Format: [DataFrameA, string]. DataFrameA: Series of Tuples(index, DataFrameB), DataFrameB: x, y.
    """
    range_ratio = min(max(0.0, range_ratio), 1.0)  # range_ratio is in the range [0.0, 1.0].
    outliers = []  # Stores the candidate outliers
    

    # Using a fixed ratio range.
    lower_bound = range_ratio * np.nanmin(data.AMOUNT)
    upper_bound = range_ratio * np.nanmax(data.AMOUNT)
    
    candidates = data.loc[(data.AMOUNT < lower_bound) | (data.AMOUNT > upper_bound)]
    outliers.append([candidates, 'Naive Range'])
    

    # Using the Modified Z-score method.
    threshold = 3.5
    ymedian = np.median(data.AMOUNT)
    ymedian_absolute_deviation = np.median(np.abs(data.AMOUNT - ymedian))
    # modified_z_score == 0.6745 * (data.y - ymedian) / ymedian_absolute_deviation
    
    candidates = []
    if ymedian_absolute_deviation != 0.0:
        candidates = data.loc[0.6745 * np.abs((data.AMOUNT - ymedian) / ymedian_absolute_deviation) > threshold]
    outliers.append([candidates, 'Modified z-Score'])

    # Using the IQR method (interquartile range).
    qr1, qr3 = np.percentile(data.AMOUNT, [25, 75])
    iqr = qr3 - qr1
    lower_bound = qr1 - iqr*1.5
    upper_bound = qr3 + iqr*1.5
    
    candidates = data.loc[(data.AMOUNT < lower_bound) | (data.AMOUNT > upper_bound)]
    outliers.append([candidates, 'Inter-Quartile Range'])


    # Group together the methods that produced the same outliers.
    if group_methods:
        unique_outliers = []
        for out in outliers:
            if not np.any([True for otl in unique_outliers if out[0].equals(otl)]):  # If the list is empty
                unique_outliers.append(out[0])
        outliers = [[otl, ', '.join([m[1] for m in outliers if m[0].equals(otl)]).strip()] for otl in unique_outliers]

    # Return format: [DataframeA, string]. DataframeA: Series of Tuples(index, DataframeB), DataframeB: x, y.
    return outliers

# display only the columns that have meaning
columns_of_interest = data[['AMOUNT', 'CLIENT_USERNAME_MASKED','DEBIT_IBAN_MASKED', 'RECEIVER_ACC_IBAN_MASKED']]

print("Outlier Detection:")

# find outliers 
outliers = find_outliers(columns_of_interest, 0.95)

# display the first 5 outliers of each method
for outlier in outliers:
    print('Method \''+outlier[1]+'\': %i Canditate(s)' %len(outlier[0]))
    display(outlier[0].head())
```

```python
# create a dictionary for the flagged accounts
flagged_by_outliers = {}

# flag each account as an outlier by the corresponding method
for i in range(len(outliers)):
    
    flagged = np.unique(np.append(outliers[i][0]['DEBIT_IBAN_MASKED'].unique(), outliers[i][0]['RECEIVER_ACC_IBAN_MASKED'].unique()))
    
    for flagged_user in flagged:
        
        if flagged_user in flagged_by_outliers.keys(): 
            flagged_by_outliers[flagged_user].append(outliers[i][1])
    
        else: 
            flagged_by_outliers[flagged_user] = [outliers[i][1]]
            
            
# calculate the probability these accounts are indeed outliers
flagged_probabilities = {}

for flagged_outlier in flagged_by_outliers.keys():
    flagged_probabilities[flagged_outlier] = len(flagged_by_outliers[flagged_outlier])/len(outliers)

# keep the accounts with probability over or equal to 0.6
outliers_flagged_accounts = list(k for k, v in flagged_probabilities.items() if v >= 0.6)

# add the flag
unique_accounts = flag_accounts(unique_accounts, outliers_flagged_accounts, 'outliers')
```

# Create Network

```python
def calculate_weights(data, weight = 'AMOUNT'):
    # put median as weight
    if weight == 'AMOUNT':
        data['from_to'] = list(zip(data['DEBIT_IBAN_MASKED'], data['RECEIVER_ACC_IBAN_MASKED']))

        sum_amount = data.groupby(['from_to'])['AMOUNT'].sum().reset_index()

        # split the tuple column (It is a messy way but I didn't know how else to do the three tuple (node_from, node_to, weight))
        sum_amount['from'], sum_amount['to'] = zip(*sum_amount.from_to)

        # make tuple with weight
        edges_weights = list(zip(sum_amount['from'], sum_amount['to'], sum_amount['AMOUNT'])) 
    
    else:
        # make tuples for nodes from and to
        data['from_to'] = list(zip(data['DEBIT_IBAN_MASKED'], data['RECEIVER_ACC_IBAN_MASKED']))

        # count the frequency an edge occurs (as a weight)
        nodes_weights = data['from_to'].value_counts().to_frame().reset_index().rename(columns = {'index': 'nodes'})

        # split the tuple column (It is a messy way but I didn't know how else to do the three tuple (node_from, node_to, weight))
        nodes_weights['from'], nodes_weights['to'] = zip(*nodes_weights.nodes)

        # make tuple with weight
        edges_weights = list(zip(nodes_weights['from'], nodes_weights['to'], nodes_weights['from_to']))
        
    return edges_weights


import networkx as nx

def create_network(data, nodes_from, nodes_to, weight = None, prints = False, graph_type = nx.Graph()):
    
    G = graph_type
    # create the graph
    G.add_weighted_edges_from(calculate_weights(data))
    
    # store the nodes and edges
    nodes, edges = list(G.nodes()), list(G.edges())
    
    # print the properties of the network
    if prints:
        
        print('The network is{} weakly connected'.format('' if nx.is_weakly_connected(G) else ' not'))
        print('The number of weakly connected components of the network are: {:,d}'.format(nx.number_weakly_connected_components(G)))
        print('The network is{} strongly connected'.format('' if nx.is_strongly_connected(G) else ' not'))
        print('The number of strongly connected components of the network are : {:,d}'.format(nx.number_strongly_connected_components(G)))

    if type(graph_type) == nx.DiGraph():
        print('The network is{} connected'.format('' if nx.is_connected(G) else ' not'))

    return (G, nodes, edges)

G, nodes, edges = create_network(data, 'DEBIT_IBAN_MASKED', 'RECEIVER_ACC_IBAN_MASKED', 'AMOUNT', graph_type = nx.DiGraph())
```

# Metrics


```python
# personalization for the pagerank
personalization = (unique_accounts[['tax_haven','outliers', 'null_reason', 'fictitious_account']].values.tolist())

personalization_dict = {}

for index in range(len(unique_accounts)):
    
    personalization_dict[unique_accounts['Account'][index]] = personalization[index]
    

from operator import itemgetter

def degree_metrics(graph, metric):
    if metric == 'in_degree': 

        degree = list(G.degree())
        degree.sort(key = itemgetter(1), reverse = True)
    elif metric == 'out_degree':

        degree = list(G.out_degree())
        degree.sort(key = itemgetter(1), reverse = True)
    degree_df = pd.DataFrame(degree).rename(columns = {0 : 'Account', 1 : metric})

    return degree_df

def centrality_metrics(graph, metric, weight = None, personalization = None):
    
    if metric == 'closeness_centrality':
        centrality = nx.closeness_centrality(graph, u = None, distance = None)
        
    elif metric == 'eigenvector_centrality':
        centrality = nx.eigenvector_centrality(graph, max_iter = 100, tol = 1e-06, nstart = None, weight = weight)
        
    elif metric == 'pagerank':
        centrality = nx.pagerank(graph, alpha=0.85, personalization = None, max_iter = 1000, tol = 1e-06, nstart = None, weight = weight, dangling = None)
        
    pd.set_option('display.float_format', '{:.5f}'.format)
    
    centrality_df = pd.DataFrame.from_dict(centrality, 'index').reset_index().rename(columns = {'index': 'Account', 0 : metric}).sort_values(by = metric, ascending = False)
    
    return centrality_df
```

```python
# calculate the metrics
pagerank = centrality_metrics(G, 'pagerank', personalization = personalization)
in_degree = degree_metrics(G, 'in_degree')
out_degree = degree_metrics(G, 'out_degree')
closeness = centrality_metrics(G, 'closeness_centrality', weight = 'weight')
eigenvector = centrality_metrics(G, 'eigenvector_centrality')


# join the basic metrics in one dataframe
joinned_df = in_degree.merge(out_degree, on = 'Account', how = 'outer').merge(pagerank, on = 'Account', how = 'outer').\
                        merge(closeness, on = 'Account', how = 'outer').merge(eigenvector, on = 'Account', how = 'outer')

# sort values by in degree ( my choice)
display(joinned_df.sort_values(by = 'in_degree', ascending = False).head(15))

# sort values by out degree
display(joinned_df.sort_values(by = 'out_degree', ascending = False).head(15))


# choose the first 50000 of each metric
top_50000_in_degree = in_degree.head(50000)
top_50000_out_degree = out_degree.head(50000)
top_50000_pagerank = pagerank.head(50000)
top_50000_closeness = closeness.head(50000)
top_50000_eigenvector = eigenvector.head(50000)

# concatenate the frames
frames = [top_50000_in_degree, top_50000_out_degree, top_50000_pagerank, top_50000_closeness, top_50000_eigenvector]
result = pd.concat(frames)

flagged_accounts = list(result.Account.unique())

# flag accounts
unique_accounts = flag_accounts(unique_accounts, flagged_accounts, 'metrics')
```

# Longest path of the graph through hash

```python
from sympy import nextprime

data['from_to'] = list(zip(data['DEBIT_IBAN_MASKED'], data['RECEIVER_ACC_IBAN_MASKED']))
# create hashes from each account and find the c value from the function (ax+b)modc where a = account_from and b = account_to and
# x is a constant
def hashed_nodes(nodes_from_to):
    
    unique_nodes = set(nodes_from_to)
    
    c = nextprime(len(unique_nodes))
    
    hashed_nodes_from_to = [[hash(account[i]) for i in range(len(account))] for account in unique_nodes]
    
    return (hashed_nodes_from_to, c)

# return the accounts and the nextprime of the length of accounts
accounts, c = hashed_nodes(data.from_to)

# create bucket table
table1 = [[] for x in range(c)]

# fill the bucket table
def buckets(nodes, hashed_nodes_from_to, c):
    
    # hash function
    index = (hashed_nodes_from_to[0]*10 + hashed_nodes_from_to[1]) % c
    
    table1[index].append(nodes)
    
    return table1



unique_nodes = list(set(data.from_to))

# fill the table with the original accounts (not the hashes) for all the nodes
for i in range(len(accounts)):
    
    buckets(unique_nodes[i], accounts[i], c)
    
    
# find the support of each group
support = [len(table1[i]) for i in range(len(table1))]

# find max support (for my usage)
#max(support)

from itertools import chain
delete_nodes = []

# find the nodes that are inside the buckets that do not exceed the minimum support
for i in range(len(table1)):
    
    if len(table1[i]) < 3:
        
        delete_nodes.append(tuple(chain(*table1[i]))) 
        
delete_observations = list(filter(None,delete_nodes))

# delete the observations with these accounts
shortened_file = data.loc[~ data.from_to.isin(delete_observations)]

# calculate the frequency of the transactions as weights
nodes_weights = calculate_weights(data, weight = 'frequency')

# remove nodes with minimum support 3
nodes_to_be_used = [weight for weight in nodes_weights if weight[2] >= 3]


# create graph
H = nx.DiGraph()
H.add_weighted_edges_from(nodes_to_be_used)
nodesH, edgesH = H.number_of_nodes(), H.number_of_edges()

# remove cycle
H.remove_edges_from(list(nx.simple_cycles(H)))


#from collections import defaultdict
from itertools import chain

def DFS(G, v, seen = None,path = None):
    if seen is None: seen = []
    if path is None: path = [v]

    seen.append(v)

    paths = []
    for t in G[v]:
        if t not in seen:
            t_path = path + [t]
            paths.append(tuple(t_path))
            paths.extend(DFS(G, t, seen[:], t_path))
    return paths

# Run DFS, compute metrics

all_paths = list(chain.from_iterable(DFS(H, n) for n in set(H)))

max_len   = max(len(p) for p in all_paths)
max_paths = [p for p in all_paths if len(p) == max_len]

source, target = [], []
for i in max_paths:
    source.append(i[0])
    target.append(i[7])
    
# these accounts can have a strange bahavior since they tranfer amounts through many other accounts.
longest_path_accounts = list(set(source + target))

# flag the above as strange important accounts
unique_accounts = flag_accounts(unique_accounts, longest_path_accounts, 'longest_path')
```

# Pagerank vs Hops 

```python
# find the flagged accounts
flagged_accounts = unique_accounts[unique_accounts.iloc[:, 1:].ne(0).any(axis=1)].reset_index(drop=True)


# number of hops
number_of_hops = [1, 2, 3, 4, 5, 6]

overall_pageranks = {}

# find the pagerank over the number of neighbors regarding hops
for node in flagged_accounts['Account']:
    
    pageranks = [] 
    
    for hop in number_of_hops:
        
        subgraph = nx.ego_graph(G, node, radius = hop)
        pageranks.append(nx.pagerank(subgraph)[node])
        
    overall_pageranks[node] = pageranks
    
# find the nodes that do not have a descending scale on their pageranks
ascending_pagerank_node = []

for node in overall_pageranks.keys():
    
    if sorted(overall_pageranks[node], reverse = True) != overall_pageranks[node]:
        ascending_pagerank_node.append(node)
        
        
# flag those accounts
unique_accounts = flag_accounts(unique_accounts, ascending_pagerank_node, 'ascending_pagerank')
```

```python
alpha = np.round(np.arange(0.01, 1, 0.01), 2)

lr = np.round(np.arange(0.0001, 0.01, 0.0001), 4)

space = {# units values
         'units1': hp.choice('units1', [40, 50, 60]),
         'units2': hp.choice('units2', [10, 20, 30]),

         # activation functions
         'activation1': hp.choice('activation1', [{'activation': 'relu'},
                                                        {'activation': 'sigmoid'},
                                                        {'activation': 'LeakyReLU',
                                                           'alpha1': hp.choice('alpha1', alpha)}]),

                 'activation2': hp.choice('activation2', [{'activation': 'relu'},
                                                          {'activation': 'sigmoid'},
                                                          {'activation': 'LeakyReLU',
                                                           'alpha2': hp.choice('alpha2', alpha)}]),

                 'activation_out': 'sigmoid',

                 # dropout values
                 'dropout1': hp.choice('dropout1', [0.12, 0.15, 0.17, 0.2, 0.22]),
                 'dropout2': hp.choice('dropout2', [0.15, 0.2, 0.25, 0.25, 0.3]),

                 # fitting parameters
                 'epochs': 100,
                 'batch_size': hp.choice('batch_size', [16, 32]),

                 'optimizer': 'Adam',
                 'lr': hp.choice('lr', lr),
                 'amsgrad' : True,

                 # EarlyStopping parameters
                 'monitor': 'val_loss',
                 'min_delta': 1e-5,
                 'patience': 50,
                 'mode': 'auto',
                 'restore_best_weights': True}
```

Undersampling

```python
def undersampling(training_data, training_target, bad_prc = 0.25, prints = False):

    bad_cases = training_target.loc[training_target == 1]
    good_cases = training_target.loc[training_target != 1]

    x_bad = training_data.iloc[bad_cases.index]

    good_cases_per_sample = int(((1 - bad_prc) * len(bad_cases)) // bad_prc)

    sample = good_cases.sample(good_cases_per_sample, replace = False, random_state = 42)

    y_resampled = pd.concat([bad_cases, sample], axis=0)

    x_resampled = pd.concat([x_bad, training_data.loc[sample.index, :]], axis=0)

    if prints:

        # Plot the distribution in the undersampled dataset
        display(pd.DataFrame(round(y_resampled.value_counts(normalize=True) * 100, 1).astype(str) + '%').rename(index={0: 'Good', 1: 'Bad'}, columns={'status': 'Undersampled'}))

    return x_resampled, y_resampled

num_classes = 2


def neural_network(training_data, training_target, scoring_metric, bad_prc = 0.25):

    def objective(params):

        np.random.seed(42)
        set_random_seed(42)

        cv_scores = []

        print('Params testing: ', params)
        print('\n ')

        cv = StratifiedKFold(n_splits = 10, shuffle = False, random_state = 42)

        for train, val in cv.split(training_data, training_target):

            model = Sequential()

            # first hidden layer and input dimension definition
            model.add(Dense(params['units1'], input_dim = training_data.shape[1], kernel_constraint = maxnorm(4)))

            if params['activation1']['activation'] == 'LeakyReLU': model.add(LeakyReLU(alpha=params['activation1']['alpha1']))

            else: model.add(Activation(params['activation1']['activation']))

            model.add(Dropout(params['dropout1']))

            # second hidden layer
            model.add(Dense(params['units2'], kernel_constraint = maxnorm(4)))

            if params['activation2']['activation'] == 'LeakyReLU':
                model.add(LeakyReLU(alpha=params['activation2']['alpha2']))

            else:
                model.add(Activation(params['activation2']['activation']))

            model.add(Dropout(params['dropout2']))

            # output layer
            model.add(Dense(num_classes - 1))
            model.add(Activation(params['activation_out']))

            # Compile model

            model.compile(loss = 'binary_crossentropy', optimizer = Adam(lr = params['lr'], amsgrad = params['amsgrad']))

            early_stopping = EarlyStopping(monitor = params['monitor'],
                                           min_delta = params['min_delta'],
                                           patience = params['patience'],
                                           restore_best_weights = params['restore_best_weights'],
                                           mode = params['mode'],
                                           verbose = 0)

            x_temp, y_temp = training_data.iloc[train].reset_index(drop = True), training_target.iloc[train].reset_index(drop = True)
            

            x_undersampled, y_undersampled = undersampling(x_temp, y_temp, bad_prc)

            model.fit(x_undersampled, y_undersampled,
                      epochs=params['epochs'],
                      batch_size=params['batch_size'],
                      callbacks=[early_stopping],
                      verbose=0)

            predictions = model.predict(training_data.iloc[val])

            if scoring_metric == 'roc_auc': 
                metric = roc_auc_score(training_target.iloc[val], predictions)

            else: 
                metric = average_precision_score(training_target.iloc[val], predictions)

            cv_scores.append(metric)

        loss = 1 - np.mean(cv_scores)

        return {'loss': loss, 'status': STATUS_OK}

    return objective
```

```python
objective = neural_network(X_train_scaled, y_training, scoring_metric = 'roc_auc', bad_prc = 0.25)
```

```python
def select_parameters(objective_function, parameter_space, max_evals, prints = True):

    trials = Trials()

    best_indices = fmin(objective_function,
                        parameter_space,
                        algo=tpe.suggest,
                        trials=trials,
                        max_evals=max_evals,
                        rstate=np.random.RandomState(42))

    best_parameters = space_eval(parameter_space, trials.argmin)

    if prints:
        print('Best params are:')
        print(best_parameters)

        main_plot_history(trials)  

    return best_parameters, best_indices, trials
```

```python
params, indices, trials = select_parameters(objective, space, 10, prints = True)
```

```python
def fit_neural_network(best_parameters, training_data, training_target):

    model = Sequential()

    # first hidden layer and input dimension definition
    model.add(Dense(best_parameters['units1'], input_dim=training_data.shape[1], kernel_constraint=maxnorm(4)))

    if best_parameters['activation1']['activation'] == 'LeakyReLU': model.add(LeakyReLU(alpha=best_parameters['activation1']['alpha1']))

    else: model.add(Activation(best_parameters['activation1']['activation']))

    model.add(Dropout(best_parameters['dropout1']))

    # second hidden layer
    model.add(Dense(best_parameters['units2'], kernel_constraint=maxnorm(4)))

    if best_parameters['activation2']['activation'] == 'LeakyReLU': model.add(LeakyReLU(alpha=best_parameters['activation2']['alpha2']))

    else: model.add(Activation(best_parameters['activation2']['activation']))

    model.add(Dropout(best_parameters['dropout2']))

    # output layer
    model.add(Dense(num_classes-1))
    model.add(Activation(best_parameters['activation_out']))

    # Compile model
    model.compile(loss='binary_crossentropy', optimizer = Adam(lr=best_parameters['lr'],
                                                                amsgrad = best_parameters['amsgrad']))

    early_stopping = EarlyStopping(monitor=best_parameters['monitor'],
                                    patience=best_parameters['patience'],
                                    restore_best_weights=best_parameters['restore_best_weights'],
                                    mode=best_parameters['mode'],
                                    verbose=0)
    # Fit the model

    history = model.fit(training_data, training_target, epochs=best_parameters['epochs'],
                                batch_size=best_parameters['batch_size'],
                                callbacks=[early_stopping], verbose=0)
    return history, model
```

```python
def plot_history(history, epochs):

    metric = 'loss'
    training_metric = history.history[metric]

    epoch_count = np.arange(1, epochs + 1)

    plt.figure(figsize=(12, 8))

    # Visualize loss history
    plt.plot(epoch_count, training_metric, 'r--', label = 'Training ' + metric.capitalize())

    plt.legend()
    plt.xlabel('Epoch')
    plt.xticks(np.arange(1, len(training_metric), 5))
    plt.ylabel(metric.capitalize())
    plt.show()
```

```python
from sklearn.metrics import roc_curve, precision_recall_curve

def plot_metrics(y_true, pred_prob, metric):

    if metric == 'auc':

        plt.figure(figsize=(8, 5))

        fpr, tpr, _ = roc_curve(y_true, pred_prob)

        plt.plot(fpr, tpr, color='orange', label = 'ROC Curve')
        plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')

        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve', fontweight='bold')

        plt.legend()

    elif metric == 'aps':

        plt.figure(figsize=(8, 5))

        precision, recall, _ = precision_recall_curve(y_true, pred_prob)

        plt.plot(recall, precision, color='orange', label='Precision-Recall')
        plt.plot([0, 1], [0.5, 0.5], color='darkblue', linestyle='--')

        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision-Recall Curve', fontweight='bold')

        plt.legend()
```
